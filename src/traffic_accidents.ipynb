5. determine the top three streets(Use the ON STREET NAME column) that had the most accidents

```python
#show the street name and the number of accidents occurred on each street

#normalizing all street names
df['OFF STREET NAME'] = df['OFF STREET NAME'].str.upper()
df['ON STREET NAME'] = df['ON STREET NAME'].str.upper()
df['CROSS STREET NAME'] = df['CROSS STREET NAME'].str.upper()


s1=df['OFF STREET NAME'].value_counts()
s2=df['ON STREET NAME'].value_counts()
s3=df['CROSS STREET NAME'].value_counts()
streets = pd.concat([s1, s2, s3])

streets.sort_values(ascending=False)
```

6. shows the number of accidents that occurred at each borough

```python
borough_count = df['BOROUGH'].value_counts(dropna=False)
borough_count
```

```python
#rename and remove NaN
borough_count['OTHER'] = borough_count[0]
borough_count = borough_count[1:]

borough_count
```

```python
#show a visualization that allows comparison of the number of the accidents
import matplotlib.pyplot as plt

plt.bar(borough_count.index, borough_count.values, 0.8)
plt.ylabel('Accidents')
plt.xlabel('Borough')
plt.title('Traffic Accidents by Borough')
plt.xticks(borough_count.index, rotation=45)
```

7. calculate summary statistics for the number of persons injured in all NYC and for a couple of selected boroughs (you can choose the two boroughs)

```python
#selecting for MANHATTAN and BROOKLYN

c1 = df[df['BOROUGH']==('MANHATTAN')]
man_injuries = c1.loc[:, ['BOROUGH', 'NUMBER OF PERSONS INJURED']].dropna()
man_injuries = man_injuries[man_injuries['NUMBER OF PERSONS INJURED'] > 0]

c2 = df[df['BOROUGH']==('BROOKLYN')]
bro_injuries = c2.loc[:, ['BOROUGH', 'NUMBER OF PERSONS INJURED']].dropna()
bro_injuries = bro_injuries[bro_injuries['NUMBER OF PERSONS INJURED'] > 0]

all_injuries = df.loc[:, ['BOROUGH', 'NUMBER OF PERSONS INJURED']].dropna()
all_injuries = all_injuries[all_injuries['NUMBER OF PERSONS INJURED'] > 0]

all_injuries.describe()
```

```python
man_injuries.describe()
```

```python
bro_injuries.describe()
```

Brooklyn has a vastly higher count of injuries, and on average more people get injured by traffic accidents in Brooklyn than they do in NYC as a whole. Brooklyn also has the highest standard deviation, and shares max values with all_injuries, which may indicate that the highest casualty count incidents in all of NYC occurred in Brooklyn. For both Brooklyn and Manhattan, however, the majority of injuries caused by traffic accidents involve just 1 person.

8. what are the distributions of accidents based on the geo location (latitude & longitude)?

```python
geo = df.loc[:, ['LATITUDE', 'LONGITUDE', 'BOROUGH']].dropna()
geo = geo[geo['LATITUDE'] > 0]
geo.describe()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

colors = {'MANHATTAN':'blue', 'BROOKLYN':'orange', 'BRONX':'green', 'STATEN ISLAND':'red', 'QUEENS':'purple'}

ax.scatter(geo['LATITUDE'], geo['LONGITUDE'], c=geo['BOROUGH'].map(colors))
```
